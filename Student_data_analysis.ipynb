# gender distribution

```python
df.head()
```

```python
plt.figure(figsize = (5,5))
sx = sns.countplot(data=df, x = "Gender")
plt.title("Gender Distribution")
sx.bar_label(sx.containers[0])
plt.show()
```

#it is clear from the above plot that number of females is more than the number of males in the given data.

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore" : 'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relation between Parent's Education and Student Scores")
plt.show()
```

```python
#from the analysis of above chart, we can conclude that the eduaction of parents highly affects the average scores of the students.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore" : 'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relation between Parent's Marital Status and Student Scores")
plt.show()
```

```python
#from the analysis of above chart, we can conclude that the marital status of parents has a no/negligible affects the average scores of the students.
```

# Distribution of Ethnic Groups

```python
ax = sns.countplot(data = df, x = 'EthnicGroup')
ax.bar_label(ax.containers[0])
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = '%1.2f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
#thus we can analyze from the above pie chart that the count of Ethnic Group C is maximum.
```
